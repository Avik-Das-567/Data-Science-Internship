# 1. Maximum Product of Two Elements in an Array

Link: https://leetcode.com/problems/maximum-product-of-two-elements-in-an-array/description/

```python
class Solution:
    def maxProduct(self, nums):
        nums.sort()
        return (nums[-1] - 1) * (nums[-2] - 1)
```

```python
# Test Result
print(Solution().maxProduct([3, 4, 5, 2]))
print(Solution().maxProduct([1, 5, 4, 5]))
print(Solution().maxProduct([3, 7]))
```

# 2. Count Number of Teams

Link: https://leetcode.com/problems/count-number-of-teams/description/

```python
class Solution:
    def numTeams(self, rating):
        n = len(rating)
        teams = 0

        for j in range(n):
            left_smaller = 0
            left_greater = 0
            right_smaller = 0
            right_greater = 0

            # check left side
            for i in range(j):
                if rating[i] < rating[j]:
                    left_smaller += 1
                elif rating[i] > rating[j]:
                    left_greater += 1

            # check right side
            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    right_greater += 1
                elif rating[k] < rating[j]:
                    right_smaller += 1

            teams += left_smaller * right_greater
            teams += left_greater * right_smaller

        return teams
```

```python
# Test Result
rating = [2, 5, 3, 4, 1]
print(Solution().numTeams(rating))
```

# 3. Number of Students Doing Homework at a Given Time

Link: https://leetcode.com/problems/number-of-students-doing-homework-at-a-given-time/description/

```python
class Solution:
    def busyStudent(self, startTime, endTime, queryTime):
        count = 0

        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1

        return count
```

```python
# Test Result
startTime = [1, 2, 3]
endTime = [3, 2, 7]
queryTime = 4

print(Solution().busyStudent(startTime, endTime, queryTime))
```

# 4. Number of Steps to Reduce a Number to Zero

Link: https://leetcode.com/problems/number-of-steps-to-reduce-a-number-to-zero/description/

```python
class Solution:
    def numberOfSteps(self, num):
        steps = 0

        while num > 0:
            if num % 2 == 0:
                num //= 2
            else:
                num -= 1
            steps += 1

        return steps
```

```python
# Test Result
print(Solution().numberOfSteps(14))
print(Solution().numberOfSteps(8))
print(Solution().numberOfSteps(123))
```

# 5. Counting Bits

Link: https://leetcode.com/problems/counting-bits/description/

```python
class Solution:
    def countBits(self, n):
        ans = [0] * (n + 1)

        for i in range(1, n + 1):
            ans[i] = ans[i // 2] + (i % 2)

        return ans
```

```python
# Test Result
print(Solution().countBits(2))
print(Solution().countBits(5))
```
